```python
# 14. Escreva um programa em linguagem C que solicite ao usurio a mdia para aprovao
# em um curso e em seguida solicite ao usurio o nome, sexo e as 03 notas do aluno e ao
# final imprima a frase: "O aluno XXXXX foi aprovado com media YY" considerando o gnero
# do(a) aluno(a) e se foi aprovado(a) ou reprovado(a)



############## POR ESTA EM C ELE SO RODA NO GDB COMPILIER ################


#include <stdio.h>
#include <string.h>

int main() {
    // Variveis para armazenar nome, gnero e notas
    char nome[50];
    char genero[10];
    int a1, a2, a3;
    float media;

    // Solicita o nome do(a) aluno(a)
    printf("Informe seu nome: ");
    scanf("%s", nome);

    // Solicita o gnero
    printf("\nMasculino\nFeminino\nInforme uma opcao acima: ");
    scanf("%s", genero);

    // Solicita as notas
    printf("Informe a nota da A1: ");
    scanf("%d", &a1);

    printf("Informe a nota da A2: ");
    scanf("%d", &a2);

    printf("Informe a nota da A3: ");
    scanf("%d", &a3);

    // Calcula a mdia
    media = (float)(a1 + a2 + a3) / 3;

    // Verifica se o(a) aluno(a) foi aprovado(a) ou reprovado(a)
    char status[20];
    if (media > 6) {
        strcpy(status, "aprovado(a)");
    } else {
        strcpy(status, "reprovado(a)");
    }

    // Imprime o resultado final
    printf("O(a) aluno(a) %s, do genero %s, foi %s com a media de %.2f\n", nome, genero, status, media);

    return 0;
}
```

```python
#15. Elabore um algoritmo que solicite N nmeros reais e quando o usurio informar o valor
#nulo 0 (zero) o programa ordene e mostre todos os nmeros informados de forma crescente.

save_num = []
print('Informe quantos numeros quiser, para encerra digite o numero \'O\' ')
print('='*60)
while True :
    num = int(input('Informe um nmero: '))
    num_real = float (num)
    if num_real == 0:
        break

    save_num.append(num_real)

if not save_num:
    print (' Nenhum numero foi inserido para ordenar')
else:
    save_num.sort()


print('='*60)
print (' Aqui esta a lista com os nmero em ordem crescente')
print('')
for fim in save_num:
    print(f'Os numeros informados foram {fim:.0f}')
```

```python
#16. Escreva um programa que v solicitando as idades dos alunos da sala at que todos
#sejam informados (perguntar ao usurio se deseja informar a idade do prximo aluno). Ao
#final apresentar a idade do mais novo, a idade do mais velho, Quantos alunos tm mais de
#18 anos, quantos alunos tm at 18 anos, a mdia aritmtica e a mediana.

# Lista para armazenar as idades
idades = []

# --- 1. Coleta das Idades (Usando 'while') ---
# O loop 'while True' vai rodar indefinidamente at o 'break'
while True:

    entrada = input("Informe a idade do aluno: ")

    # Condio de parada, e garantir que a idade  um nmero positivo
    if entrada.lower() == 'parar':
        break

    try:
        idade = int(entrada)
        print("Digite 'parar' quando terminar.")
        if idade > 0 and idade <= 120:
            idades.append(idade)
        else:
            print("Entrada Indevida. Por favor, insira uma idade valida at 120 anos.")
    except ValueError:
        print("Entrada invlida. Por favor, insira um nmero ou 'parar'.")

# Varivel para contar o nmero total de alunos
total_alunos = 0

# Usei o 'for' para contar o total (substitui len(idades))
for idade in idades:
    total_alunos += 1

# Se a lista estiver vazia, encerra apos digitar 'parar'
if total_alunos == 0:
    print("\nNenhuma idade foi informada. Encerrando o programa.")

    # Usando a funo 'exit()' para parar o sistema sem a necessidade de uma funo
    exit()


# Mais Novo e Mais Velho (Substituindo min/max)
idade_mais_novo = idades[0]  # Assume que o primeiro  o menor
idade_mais_velho = idades[0] # Assume que o primeiro  o maior

# Loop 'for' para encontrar o menor e o maior
for idade in idades:
    if idade < idade_mais_novo:
        idade_mais_novo = idade
    if idade > idade_mais_velho:
        idade_mais_velho = idade

#Contagem de Alunos (>18 e <=18)
cont_maiores_18 = 0
cont_menores_ou_igual_18 = 0

# Loop 'for' para contagem
for idade in idades:
    if idade > 18:
        cont_maiores_18 += 1
    else:
        cont_menores_ou_igual_18 += 1

# Mdia Aritmtica
soma_idades = 0

# Loop 'for' para somar todas as idades
for idade in idades:
    soma_idades += idade


# Mediana (Substituindo sorted())
media_aritmetica = soma_idades / total_alunos

# Implementao manual do Bubble Sort (o mtodo de ordenao mais simples)
n = total_alunos

# Uma cpia da lista para ordenao (para no alterar a original)
idades_ordenadas = idades[:]

# Dois loops 'for' aninhados para ordenar a lista
for i in range(n):
    for j in range(0, n-i-1):
        if idades_ordenadas[j] > idades_ordenadas[j+1]:
            # Troca de posio
            temp = idades_ordenadas[j]
            idades_ordenadas[j] = idades_ordenadas[j+1]
            idades_ordenadas[j+1] = temp

# Clculo da Mediana
if n % 2 == 1:
    # mpar: Mediana  o elemento do meio (o ndice  n // 2)
    mediana = idades_ordenadas[n // 2]
else:
    # Par: Mediana  a mdia dos dois elementos do meio
    # Os ndices so n // 2 - 1 e n // 2
    meio_1 = idades_ordenadas[n // 2 - 1]
    meio_2 = idades_ordenadas[n // 2]
    mediana = (meio_1 + meio_2) / 2

# --- 3. Apresentao dos Resultados ---

print("\n" + "="*40)
print("           RESULTADOS FINAIS")
print("="*40)
print(f"Total de alunos informados: {total_alunos}")
print("-" * 40)

print(f"Idade do Mais Novo: {idade_mais_novo} anos")
print(f"Idade do Mais Velho: {idade_mais_velho} anos")

print("-" * 40)

print(f"Alunos com mais de 18 anos: {cont_maiores_18}")
print(f"Alunos com at 18 anos: {cont_menores_ou_igual_18}")

print("-" * 40)

print(f"A mdia aritmtica das idades: {media_aritmetica:.2f} anos")
print(f"A mediana das idades: {mediana:.2f} anos")
print("="*40)
```
